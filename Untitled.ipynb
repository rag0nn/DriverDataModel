```python
import pandas as pd
import numpy as np
from keras.layers import Dense,Input,Dropout
from keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Drivers_Data.csv")
df
```

```python
df.columns
```

```python
df = df.drop(columns=["Driver_ID","Name"])
```

```python
le = LabelEncoder()
df["City"] = le.fit_transform(df["City"])
df
```

```python
df["Active_Status"] = df["Active_Status"].map({"Active":1,"Inactive":0})
```

```python
df
```

```python
df.describe()
```

```python
df = df.sort_values(by=["Age"])
```

```python
import matplotlib.pyplot as plt
df.plot(x="Age",y="Average_Rating",kind='line',title="Age-Rating")
plt.xlabel("Age")
plt.ylabel("Rating")
```

```python
df.plot(kind="box",)
```

## Modelleme

```python
model = Sequential()
model.add(Dense(5))
model.add(Dropout(0.2))
model.add(Dense(5))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
from keras.losses import 
```

```python
model.compile(optimizer='Adam',metrics=["Accuracy","mse"],loss="mse")
```

```python
hist = model.fit(x=df[["Age","City","Experience_Years","Average_Rating"]][:90],
         y=df["Active_Status"][:90],
         epochs=1000,
         validation_split=0.1
         )
```

```python
hist.history.keys()
```

```python
plt.plot(range(len(hist.history["Accuracy"])),hist.history["Accuracy"])
```

```python
plt.plot(range(len(hist.history["mse"])),hist.history["mse"])
```

```python
plt.plot(range(len(hist.history["val_Accuracy"])),hist.history["val_Accuracy"])
```

```python
plt.plot(range(len(hist.history["val_mse"])),hist.history["val_mse"])
```

```python
evaluations = model.evaluate(x=df[["Age","City","Experience_Years","Average_Rating"]][90:],
         y=df["Active_Status"][90:])
```

```python
evaluations
```

```python
prediction = model.predict(x=np.array([29,1,10,4]).reshape(-1,4))

print(prediction)
treshold = 0.5
if prediction < treshold:
    print("Inactive")
else:
    print("Active")
```
